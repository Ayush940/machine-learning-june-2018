```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=1000, factor=0.4)
```

```python
for ix in range(X.shape[0]):
    if y[ix] == 0:
        plt.scatter(X[ix, 0], X[ix, 1], c='r')
    else:
        plt.scatter(X[ix, 0], X[ix, 1], c='b')

plt.show()
```
